# Import Libraries

```python
import os
import time
import joblib
import numpy as np
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from lightgbm import LGBMClassifier
from sklearn.metrics import accuracy_score
```

# Load Data Training

```python
# Load pre-extracted HOG features and encoded labels
X_train = np.load('data/features_hog/X_train_hog.npy')
X_val = np.load('data/features_hog/X_val_hog.npy')
y_train = np.load('data/processed/y_train.npy')
y_val = np.load('data/processed/y_val.npy')

print(f"Train Set | X: {X_train.shape}, y: {y_train.shape}")
print(f"Val Set   | X: {X_val.shape}, y: {y_val.shape}")
```

# Train SVM

SVM functions as the primary architectural model for this spatial trajectory task due to its mathematical synergy with high-dimensional feature representations like HOG.

```python
print("=== Training Support Vector Machine ===")
start_time = time.time()

svm_model = SVC(kernel='rbf', C=10, gamma='scale', random_state=42)
svm_model.fit(X_train, y_train)
svm_time = time.time() - start_time

y_pred_svm = svm_model.predict(X_val)
svm_acc = accuracy_score(y_val, y_pred_svm)

output_dir = 'model_saved'
os.makedirs(output_dir, exist_ok=True)
joblib.dump(svm_model, os.path.join(output_dir, 'svm_hangul_model.pkl'))

print(f"SVM Training Time : {svm_time:.2f} seconds")
print(f"SVM Val Accuracy  : {svm_acc * 100:.2f}%")
```

# Train KNN

KNN serves as a baseline, instance-based learning model to evaluate character classification purely through spatial proximity metrics.

```python
print("=== Training K-Nearest Neighbors ===")
start_time = time.time()

knn_model = KNeighborsClassifier(n_neighbors=5, weights='distance', n_jobs=-1)
knn_model.fit(X_train, y_train)
knn_time = time.time() - start_time

y_pred_knn = knn_model.predict(X_val)
knn_acc = accuracy_score(y_val, y_pred_knn)

joblib.dump(knn_model, os.path.join(output_dir, 'knn_hangul_model.pkl'))

print(f"KNN Training Time : {knn_time:.2f} seconds")
print(f"KNN Val Accuracy  : {knn_acc * 100:.2f}%")
```

# Train LightGBM

LightGBM represents tree-based ensemble learning, utilizing a leaf-wise tree growth strategy. It is introduced as a high-performance alternative optimized for real-time edge processing.

```python
print("=== Training LightGBM ===")
start_time = time.time()

lgb_model = LGBMClassifier(n_estimators=100, random_state=42, n_jobs=-1, verbosity=-1)
lgb_model.fit(X_train, y_train)
lgb_time = time.time() - start_time

y_pred_lgb = lgb_model.predict(X_val)
lgb_acc = accuracy_score(y_val, y_pred_lgb)

joblib.dump(lgb_model, os.path.join(output_dir, 'lgbm_hangul_model.pkl'))

print(f"LightGBM Training Time : {lgb_time:.2f} seconds")
print(f"LightGBM Val Accuracy  : {lgb_acc * 100:.2f}%")
```
